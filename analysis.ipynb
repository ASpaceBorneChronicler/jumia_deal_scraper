```python
import pandas as pd
import re
```

```python
df = pd.read_csv('jumia_flash_sales.csv')
df
```

```python
df['price'] = df['price'].str.replace('KSh ','')
df['old price'] = df['old price'].str.replace('KSh ','')
df
```

```python
discounted_df =df[df['old price'].notna()]
discounted_df
```

```python
discounted_rated_df = discounted_df[discounted_df['rating'] != 'No rating']
discounted_rated_df
```

```python
rating = discounted_rated_df['rating'].str.extract(r'(\d+(\.\d+)?)')
cleaned_df = discounted_rated_df.copy()
cleaned_df.drop('rating', axis=1, inplace=True)
cleaned_df['rating'] = rating[0]
cleaned_df
```

```python
cleaned_df.dtypes
```

```python
cleaned_df.price = cleaned_df.price.str.replace(',','')
cleaned_df['old price'] = cleaned_df['old price'].str.replace(',','')
```

```python
cleaned_df.price =cleaned_df.price.astype(float)
cleaned_df['old price']=cleaned_df['old price'].astype(float)
cleaned_df.rating=cleaned_df.rating.astype(float)
cleaned_df.dtypes
```

```python
cleaned_df['perc_dif']=((cleaned_df['old price']-cleaned_df['price'])/cleaned_df['old price'])*100
cleaned_df
```

```python
cleaned_df.perc_dif=cleaned_df.perc_dif.round(2)
cleaned_df
```

```python
cleaned_df.sort_values(by='perc_dif', ascending=False)
```

```python
cleaned_df.to_csv('jumia_flash_sales-clean.csv', index=False)
```
